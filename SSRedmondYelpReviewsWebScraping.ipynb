```python
import requests
from bs4 import BeautifulSoup
from urllib.request import urljoin
import pandas as pd
import re
import time
import numpy as np
```

```python
reviews_data = []
id_num = 0
url_start = "https://www.yelp.com/biz/silver-spoon-thai-restaurant-redmond"
response = requests.get(url_start)
yelp = response.content
yelpSoup = BeautifulSoup(yelp)

pages_text = yelpSoup.find("div",class_="border-color--default__09f24__NPAKY text-align--center__09f24__fYBGO").text
num_pages = pages_text.split()[2]
num_pages = int(num_pages)

pages = list(range(1,num_pages+1))
pages_groups = np.array_split(np.array(pages),4)
```

```python
## Pages Group 1 of 4

pages_group1 = pages_groups[0]
reviews_data_group1 = []

for page in pages_group1:
    
    if page != 1:
        url = url_start+"?start="+str((page*10)-10)
        response_group1 = requests.get(url)
        time.sleep(3)
        yelp = response_group1.content
        yelpSoup = BeautifulSoup(yelp)

    DIV_text = yelpSoup.find("div",{"class": "css-79elbk border-color--default__09f24__NPAKY", "aria-live": "polite", "role":"region"})
    DIV_border = DIV_text.find_all("div")[0]
    DIV_uls = DIV_border.find("ul", class_ = "undefined list__09f24__ynIEd")
    DIV_reviews = DIV_uls.find_all("li")
    num_reviews = len(DIV_reviews)

    for review in range(num_reviews):
        id_num = id_num+1
        review_LI = DIV_reviews[review]
        review_DIV = review_LI.find_all("div")[0]

        ## Gets reviewer info
        reviewer_info_DIV = review_DIV.find("div")
        reviewer_info = reviewer_info_DIV.text
        reviewer_info_split = reviewer_info.split('.')
        if len(reviewer_info_split) == 3: 
            # Gets name
            reviewer_name = reviewer_info_split[0]+"."+reviewer_info_split[1]+"."
            # Gets location
            reviewer_location = re.split(r'(\d+)',reviewer_info_split[2])
            reviewer_location = reviewer_location[len(reviewer_location)-3]
            reviewer_location_split = reviewer_location.split(", ")
            reviewer_city = reviewer_location_split[0]
            reviewer_state = reviewer_location_split[1]
        else:
            # Gets name
            reviewer_name = reviewer_info_split[0]+"."
            # Gets location
            reviewer_location = re.split(r'(\d+)',reviewer_info_split[1])
            reviewer_location = reviewer_location[len(reviewer_location)-3]
            reviewer_location_split = reviewer_location.split(", ")
            reviewer_city = reviewer_location_split[0]
            reviewer_state = reviewer_location_split[1]

        ## Gets reviewer rating info
        rating_DIV = review_DIV.find("div",class_="margin-t1__09f24__w96jn margin-b1-5__09f24__NHcQi border-color--default__09f24__NPAKY")
        # Gets reviewer rating and date
        rating_DIV_layer1 = rating_DIV
        rating_DIV_layer2 = rating_DIV_layer1.find("div")
        review_date = rating_DIV_layer1.find_all("span")[1].text
        rating_DIV_layer3 = rating_DIV_layer2.find("span")
        rating_DIV_layer4 = rating_DIV_layer3.find("div")
        rating_text_split = rating_DIV_layer4['aria-label'].split(" ")
        rating_num = rating_text_split[0]

        ## Gets review text
        review_text_DIV = review_DIV.find("p",class_="comment__09f24__gu0rG css-1sufhje")
        review_text = review_text_DIV.text
    
        ## Adds to reviews_data 
        reviews_data_group1.append([id_num,reviewer_name,reviewer_city,reviewer_state,review_date,rating_num,review_text])
    
        print(reviewer_name)
        print(id_num)
        #print(id_num)
        #print(reviewer_city)
        #print(reviewer_state)
        #print(review_date)
        #print(rating_num)
        #print(review_text)
        #print("")
        
    time.sleep(3)

time.sleep(60)
```
